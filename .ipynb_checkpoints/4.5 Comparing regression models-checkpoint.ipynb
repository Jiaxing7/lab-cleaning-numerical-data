```python
import pandas as pd
import numpy as np 
```

```python
Numerical = pd.read_csv('numerical.csv')
Numerical
```

```python
Categorical = pd.read_csv('categorical.csv')
Categorical
```

```python
data = pd.concat((Numerical, Categorical), axis=1)
data
```

```python
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower().replace(' ', '_'))
data.columns = cols

print(data.columns)
```

### 1.In this final lab, we will model our data. Import sklearn train_test_split and separate the data.

```python
X = data.drop(['total_claim_amount'],axis=1)
y = data['total_claim_amount']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
X_train.shape
```

### 2.Separate X_train and X_test into numerical and categorical (X_train_cat , X_train_num , X_test_cat , X_test_num)

```python
X_train_num = X_train.select_dtypes(np.number)
X_test_num = X_test.select_dtypes(np.number)
X_test_num.head()
```

```python
X_train_cat= X_train.select_dtypes(object)
X_test_cat = X_test.select_dtypes(object)
X_train_cat .head()
```

### 3.Use X_train_num to fit scalers. Transform BOTH X_train_num and X_test_num.

```python
from sklearn.preprocessing import StandardScaler

transformer = StandardScaler().fit(X_train_num)
numericals_train_standardized = transformer.transform(X_train_num)
numericals_test_standardized = transformer.transform(X_test_num)
```

```python
pd.DataFrame(numericals_train_standardized).head()
```

### 4.Encode the categorical variables X_train_cat and X_test_cat (See the hint below for encoding categorical data!!!)

```python
onehot_train_cat = X_train_cat[['state', 'marital_status', 'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class']]
onehot_test_cat = X_test_cat[['state', 'marital_status', 'policy_type', 'policy', 'renew_offer_type', 'sales_channel', 'vehicle_class']]
onehot_test_cat
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(handle_unknown='error', drop='first').fit(onehot_train_cat)
categoricals_train_encoded = encoder.transform(onehot_train_cat).toarray()
categoricals_test_encoded = encoder.transform(onehot_test_cat).toarray()
```

```python
encoder.categories_
```

```python
categoricals_test_encoded
```

```python
ordinal_train_cat = X_train_cat[['coverage', 'employmentstatus', 'location_code', 'vehicle_size']]
ordinal_test_cat = X_test_cat[['coverage', 'employmentstatus', 'location_code', 'vehicle_size']]
ordinal_test_cat
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal_encoder = OrdinalEncoder()
# fit_transform on the specified columns
ordinal_test_cat = ordinal_encoder.fit_transform(X_test_cat[['coverage', 'employmentstatus', 'location_code', 'vehicle_size']])
ordinal_train_cat = ordinal_encoder.fit_transform(X_train_cat[['coverage', 'employmentstatus', 'location_code', 'vehicle_size']])
ordinal_test_cat
```

```python
X_train = np.concatenate((numericals_train_standardized,categoricals_train_encoded,ordinal_train_cat),axis=1)
X_test = np.concatenate((numericals_test_standardized,categoricals_test_encoded,ordinal_test_cat),axis=1)
```

```python
pd.DataFrame(X_train)
```

### 7.Now define a function that takes a list of models and train (and tests) them so we can try a lot of them without repeating code.

```python
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor

LR = linear_model.LinearRegression()
LR.fit(X_train, y_train)
Lscore_train = LR.score(X_train, y_train)
Lscore_test = LR.score(X_test, y_test)
print('LinearRegressor score train is ', Lscore_train)
print('LinearRegressor score test is ', Lscore_test)


KNN = KNeighborsRegressor(n_neighbors=6)
KNN.fit(X_train, y_train)
Kscore_test = KNN.score(X_test, y_test)
Kscore_train = KNN.score(X_train, y_train)
print('KneighborsRegressor score train is ', Kscore_train)
print('KneighborsRegressor score test is ', Kscore_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor

def train_regression_models(models, test_size=0.2, random_state=42):
    results = {}
    for name, model in models.items():
        score_train = model.score(X_train, y_train)
        score_test = model.score(X_test, y_test)
        print(f"Train {name}:{score_train} ")
        print(f"Test {name}:{score_test} ")
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        r2 = r2_score(y_test, y_pred)
        results[name] = r2
        print(f"R2 score of {name}: {r2}")
    return results
```
